```python
import duckdb
import pandas as pd
import os
from datetime import datetime
```

```python
con = duckdb.connect('dados_duckdb.db', read_only=False)
```

```python
arquivo = "z0019_2.csv"
data_ingestao = datetime.now()

df = pd.read_csv(f'../landing/{arquivo}', sep=";")

df['nome_arquivo'] = arquivo
df['data_ingestao'] = data_ingestao

df.head()
```

```python
con.execute("""
    CREATE TABLE IF NOT EXISTS bronze_produtos (
        NATBR VARCHAR,
        MAKTX VARCHAR,
        WERKS VARCHAR,
        MAINS VARCHAR,
        LABST VARCHAR,
        nome_arquivo VARCHAR,
        data_ingestao TIMESTAMP
    )
""")
```

```python
con.execute("""
    INSERT INTO bronze_produtos SELECT * FROM df
""")
```
